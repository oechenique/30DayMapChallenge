```python
# Define the region of interest for Fukushima
fukushima_region = ee.Geometry.Point([141.032, 37.421]).buffer(20000)  # 20 km buffer around the Fukushima nuclear plant

# Function to get an annual quality composite image with a quality filter
def get_quality_composite(year, num_years=1):
    start_date = f'{year}-01-01'
    end_date = f'{year + num_years - 1}-12-31'

    # Landsat 8 collection (use Landsat 5 for earlier years if needed)
    collection = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
                 .filterBounds(fukushima_region) \
                 .filterDate(start_date, end_date) \
                 .filterMetadata('CLOUD_COVER', 'less_than', 20)

    # Quality mask based on NDVI to remove images with low vegetation (indicating clouds or shadows)
    def mask_quality(image):
        ndvi = image.normalizedDifference(['SR_B5', 'SR_B4']).rename('NDVI')
        mask = ndvi.gt(0.1)  # Filter out pixels with low NDVI, indicating shadows or clouds
        cloud_mask = image.select('QA_PIXEL').bitwiseAnd(int('11111', 2)).eq(0)
        return image.updateMask(cloud_mask.And(mask))

    # Apply the quality mask and calculate the median of the period
    composite = collection.map(mask_quality).median()
    return composite.multiply(0.0000275).add(-0.2).visualize(bands=['SR_B4', 'SR_B3', 'SR_B2'], min=0, max=0.3)

# Create a list of high-quality composite images for each year
years = list(range(2010, 2023))  # From 2011 to capture changes after the nuclear disaster
images = [get_quality_composite(year, num_years=2).set('label', str(year)) for year in years]

# Set up the output directory for the GIF
import os
out_dir = os.path.join(os.path.expanduser("~"), "Downloads")
if not os.path.exists(out_dir):
    os.makedirs(out_dir)

# Generate the GIF
Map = geemap.Map(center=[37.421, 141.032], zoom=10)
label = "Post-Disaster Changes in Fukushima, Japan"
Map.add_landsat_ts_gif(
    roi=fukushima_region,
    out_gif=os.path.join(out_dir, "fukushima_timelapse.gif"),
    start_year=2010,
    end_year=2023,
    bands=["Red", "Green", "Blue"],
    font_color="white",
    frames_per_second=2,
    font_size=30,
    progress_bar_color="blue",
    label=label,
    apply_fmask=True  # Keep the cloud mask applied
)
```
